# NASA Mars News

```python
url='https://mars.nasa.gov/news/'
```

```python
# # Retrieve page with the requests module
response = requests.get(url)
```

```python
# Create BeautifulSoup object; parse with 'html.parser'
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
# Examine the results, then determine element that contains sought info
# print(soup.prettify())
```

```python
results = soup.find_all('div', class_="slide")

first_title=[]
first_p=[]
first_link=[]

# Loop through returned results
for result in results:
    # Error handling
    try:
        # Identify and return title 
        title = result.find('div', class_='content_title').text
        first_title.append(title)
        
        #Identify and return paragraph
        p = result.find('div', class_='rollover_description_inner').text
        first_p.append(p)
        
        #identify and return link
        link = result.a['href']
        first_link.append(link)

    except AttributeError as e:
        print(e)

first_title = first_title[0]
first_p = first_p[0]
first_link = first_link[0]
        
print(first_title)
print(first_p)
print(first_link)
```

# PL Mars Space Images - Featured Image

```python
# Setup splinter
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
url = 'https://spaceimages-mars.com/'
browser.visit(url)
```

```python
html = browser.html
soup = BeautifulSoup(html, 'html.parser')
print(soup.prettify())
```

```python
results = soup.find_all('div', class_="floating_text_area")
print(results)
```

```python
# Loop through returned results
for result in results:
    # Error handling
    try:
        # Identify and return image
        featured = result.a['href']
        featured_image_url = "https://spaceimages-mars.com/" + featured
        

    except AttributeError as e:
        print(e)

print(featured_image_url)
```

# Mars Hemispheres

```python
# Setup splinter
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
url = 'https://marshemispheres.com/'
browser.visit(url)
```

```python
# Example:
hemisphere_image_urls = []
```

```python
browser.click_link_by_partial_text('Cerberus')

html = browser.html
    # Parse HTML with Beautiful Soup
soup = BeautifulSoup(html, 'html.parser')

# print(soup.prettify())

cerberus_title = soup.find('h2', class_="title").text

cerberus_img_url = soup.find(class_="downloads").a['href']

dict_1 = {"title":cerberus_title, "img_url": cerberus_img_url}
hemisphere_image_urls.append(dict_1)
```
